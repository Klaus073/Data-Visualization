```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random 
```

```python
import pandas as pd
import random

# Create a dictionary with the data
data = {
    'Date': [],
    'Total': [],
    'Accept': [],
    'Reject': []
}

# Generate data for each month
months = ['May', 'June', 'July']
for month in months:
    # Generate a sequence of dates for the month
    dates = pd.date_range(start=f'2023-{month}-01', periods=30, freq='D')

    total_values = random.sample(range(235, 301), 30)
    accept_values = random.sample(range(200, 291), 30)
    reject_values = [max(t - a, 0) for t, a in zip(total_values, accept_values)]
    
    data['Date'].extend(dates)
    data['Total'].extend(total_values)
    data['Accept'].extend(accept_values)
    data['Reject'].extend(reject_values)

# Create the DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
total1 = random.sample(range(270, 300), 30)
accept1 = random.sample(range(230, 280), 30)

total2 = random.sample(range(270, 300), 30)
accept2 = random.sample(range(230, 270), 30)

total3 = random.sample(range(270, 300), 30)
accept3 = random.sample(range(230, 260), 30)
```

```python
# Create the first DataFrame for 'Total' values
df_total = pd.DataFrame({'Total1': total1, 'Total2': total2, 'Total3': total3})

# Create the second DataFrame for 'Accept' values
df_accept = pd.DataFrame({'Accept1': accept1, 'Accept2': accept2, 'Accept3': accept3})

# Create the third DataFrame for 'Reject' values
df_reject = pd.DataFrame({
    'Reject1': [max(t - a, 0) for t, a in zip(total1, accept1)],
    'Reject2': [max(t - a, 0) for t, a in zip(total2, accept2)],
    'Reject3': [max(t - a, 0) for t, a in zip(total3, accept3)]
})

# Print the DataFrames
print("Total DataFrame:")
print(df_total)

print("\nAccept DataFrame:")
print(df_accept)

print("\nReject DataFrame:")
print(df_reject)
```

```python
def line_plot(total,data,xlabel,ylabel,title,a_r):

    sns.lineplot(data=data ,marker='o')
    per=sum(data)/sum(total)*100
    # Set labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(f'{a_r} Sample data for {title} : {per}')

    # Display the plot
    plt.show()
    
#     print(total)
#     print(data)
```

```python
def box_plot(total,data,xlabel,ylabel,title,a_r):
    # Create the box plot
    sns.boxplot(data=data)
    
    per=sum(data)/sum(total)*100

    # Set labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(f'{a_r} Sample data for {title} : {per}')

    # Display the plot
    plt.show()
    
   
    
```

```python
def box_plot_mul(df):
    
   
    
    sns.boxplot(data=df)
```

```python
def monthly_data(total,data, x_label , y_label , month  ):
    
    
    reject = [max(t - a, 0) for t, a in zip(total, data)]

    # Create a DataFrame from the data
    df = pd.DataFrame({'Total': total, 'Accept': data, 'Reject': reject})
    
    
    line_plot(df['Total'],df['Accept'], x_label , y_label , month,"Accept")
    line_plot(df['Total'],df['Reject'], x_label , y_label , month , "Reject")
    
    box_plot(df['Total'],df['Reject'], x_label , y_label , month , "Reject")
    box_plot(df['Total'],df['Accept'], x_label , y_label , month,"Accept")
    
```

```python
monthly_data(total1,accept1,"Day","Samples","May")
monthly_data(total2,accept2,"Day","Samples","June")
monthly_data(total2,accept2,"Day","Samples","July")


```
