```python
male = df[df['Sex'] == 'male']
male_die= male[male['Survived'] == 0]

female = df[df['Sex'] == 'female']
female_die= female[female['Survived'] == 0]

```

```python
Sex = ['Male', 'Female']
male_female_deaths = []
male_female_deaths.append(len(male_die))
male_female_deaths.append(len(female_die))
```

```python
#bar plot in Matplotlib

total = sum(male_female_deaths)
plt.title("Male And Female Died in Titanic")

plt.bar(Sex,male_female_deaths)
for i, v in enumerate(male_female_deaths):
    plt.text(i, v + 1, f'{(v/total)*100:.1f}%', ha='center')

plt.show()
```

```python
#bar plot in Seaborn

colors = ["#ff7f0e", "#2ca02c"]
sns.barplot(x=Sex, y=male_female_deaths , palette = colors)
```
